**Задание 2**<br/>
Используем файл keywords.csv.<br/>
- оценка 2 и меньше - низкий рейтинг<br/>
- оценка 4 и меньше - средний рейтинг<br/>
- оценка 4.5 и 5 - высокий рейтинг<br/>

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
data = pd.read_csv('keywords.csv')
data.head()
```

```python
max_val_shows = data['shows'].max()
max_val_shows
```

```python
def classification(row):
    if (row/max_val_shows)*5 <= 2:
        return 'низкий рейтинг'
    elif 2 < (row/max_val_shows)*5 <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
data['class'] = data['shows'].apply(classification)
new_data = data
new_data.head()
```

**Задание 2**<br/>
Используем файл keywords.csv.<br/>

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.<br/>

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:<br/>

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'cахалин', 'хабаровск']
}
```

Результат классификации запишите в отдельный столбец region.

```python
df = pd.DataFrame(
    {'keyword': ['Тула клуб', 'jlyjrkfcy', 'вк', 'cахалин кафе', 'vvvvv'], 'shows' : [11111, 55555, 66666, 99999, 88888888], 
     'class' : ['ggggg', 'weweee', 'oooooo', 'qqqqqq', 'xxxxxxx']}
)

df
```

```python
def geo_classification(row):
    for key, value in geo_data.items():
        for val in value:
            if str(val.upper()) in str(row['keyword']).upper():
                return key
    return 'undefined'
```

```python
df['region'] = df.apply(geo_classification, axis=1)
df
```

```python
new_data['region'] = new_data.apply(geo_classification, axis=1)
new_data
```

```python
new_data = new_data[ (new_data['keyword'].str.contains('тула', case = True)) ]
new_data
```

**Задание 3 (бонусное)**<br/>
Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.<br/>

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:<br/>

1. В переменную years запишите список из всех годов с 1950 по 2010.<br/>

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:<br/>

  1. для каждой строки пройдите по всем годам списка years
  2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
  3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
  
3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

Важно! При копировании юпитер ноутбука изменяет кавычки ’ на ` , что может способствовать ошибки в коде. В данной ситуации необходимо их просто заменить.

```python
ratings = pd.read_csv('ml-latest/ratings.csv')
```

```python
movies = pd.read_csv('ml-latest/movies.csv')
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
```

```python
years = []
for n in range(1950,2011):
    years.append(n)
print(years, sep = '')
```

```python
def production_year(row):
    for yr in years:
        if str(yr) in row:
            return yr
    return 1900
```

```python
joined['year'] = joined['title'].apply(production_year)
```

```python
joined.groupby('userId')
```

```python
new_joined = joined.groupby('year').mean()[['rating']].sort_values(by='rating', ascending=False).head(100)
new_joined
```
